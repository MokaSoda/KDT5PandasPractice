(1) 모듈 로딩


```python

from pandas import DataFrame, read_csv
import random
```

(2) 데이터 준비 및 pandas 형태로 저장

```python
movieDF = read_csv('../DATA/movies.csv')
```

(4) DataFrame의 속성들 출력

```python
movieDF.index, movieDF.columns, movieDF.values
```

(5) DataFrame의 컬럼별로 따로 저장하기

```python
movieDFcolumns = []
for x in movieDF.columns:
    movieDFcolumns.append(movieDF[x])
```

(6) 컬럼별 데이터들의 속성 즉, index, values, shape, ndim, dtype 정보 출력

```python
movieDFcolumnsValues = []
for x in movieDFcolumns:
    print('인덱스 :', x.index, 'shape :', x.shape, '차원 :', x.ndim, '데이터 타입 :',  x.dtype)
    movieDFcolumnsValues.append(x.values)
    
print(movieDFcolumnsValues)
```

(7) 컬럼별 인덱스를 변경 실습 

```python
# using rank info
rank = movieDFcolumns.pop(0)
ranklist = list(rank)
```

```python
# 순위를 인덱스로 잡아 모두 집어 넣습니다. 
movieDFcolumns_with_rankindex = []
for x in movieDFcolumns:
    x.index = rank
    movieDFcolumns_with_rankindex.append(x)
    
```

```python
# 랭크가 index로 지정된 상태로 모든 다른 Series이 같은 index를 가지게 되었습니다.
# 따라서 순위만 입력하면 해당 영화의 properties를 보여줍니다.
targetRank = 15
for x in movieDFcolumns_with_rankindex:
    print(x.loc[targetRank])
```

(8) 컬럼별 데이터 즉 값 변경

```python
target = random.choice(movieDF.columns)
print(target)

for x in movieDF.index:
    movieDF[target].iloc[int(x)-1] = 'Censored'
```

(9) DataFrame의 원소들 모두 출력 단, 2가지 방법 사용

```python
# loc 을 활용한 방법
for x in movieDF.index:
    for y in movieDF.columns:
        print(movieDF.loc[x, y])
```

```python
# iloc을 사용하는 방법
for x in range(movieDF.shape[0]):
    for y in range(movieDF.shape[1]):
        print(movieDF.iloc[x,y])
```
